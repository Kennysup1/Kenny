```python
import nltk
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')
print("NLP libraries ready! ✅")
```

```python
sentences = [
    "I love this product, it is amazing!",
    "This is the worst experience I have ever had.",
    "The weather today is okay, nothing special.",
    "I am so happy with my new laptop!",
    "This movie was absolutely terrible and boring.",
    "The food was decent but could be better.",
    "Incredible service, I will definitely come back!",
    "I hate waiting in long queues, very frustrating."
]

results = []
for sentence in sentences:
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        sentiment = "Positive 😊"
    elif polarity < 0:
        sentiment = "Negative 😞"
    else:
        sentiment = "Neutral 😐"
    results.append({'Text': sentence[:40]+'...', 'Polarity': polarity, 'Sentiment': sentiment})

df = pd.DataFrame(results)
print(df)
```

```python
colors = ['green' if p > 0 else 'red' if p < 0 else 'gray' 
          for p in df['Polarity']]

plt.figure(figsize=(10, 6))
plt.barh(range(len(df)), df['Polarity'], color=colors)
plt.yticks(range(len(df)), [f"Sentence {i+1}" for i in range(len(df))])
plt.xlabel('Polarity Score (-1 negative to +1 positive)')
plt.title('Sentiment Analysis - Kenny NLP')
plt.axvline(x=0, color='black', linewidth=0.5)
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
my_text = "I am really enjoying learning AI and ML with Kenny!"
blob = TextBlob(my_text)
print(f"Text: {my_text}")
print(f"Polarity: {blob.sentiment.polarity:.2f}")
print(f"Subjectivity: {blob.sentiment.subjectivity:.2f}")
if blob.sentiment.polarity > 0:
    print("Sentiment: Positive 😊")
elif blob.sentiment.polarity < 0:
    print("Sentiment: Negative 😞")
else:
    print("Sentiment: Neutral 😐")
```

```python
reviews = [
    "This phone is absolutely amazing, best purchase ever!",
    "Terrible quality, broke after one day. Waste of money!",
    "Decent product, works as expected nothing special.",
    "I love everything about this, highly recommend!",
    "Very disappointed, does not match the description at all.",
    "Great value for money, very happy with my purchase!",
    "Awful customer service, never buying from here again.",
    "Perfect gift, my friend loved it so much!"
]

print("=" * 60)
print("KENNY PRODUCT REVIEW ANALYZER")
print("=" * 60)

positive, negative, neutral = 0, 0, 0

for i, review in enumerate(reviews):
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        sentiment = "Positive 😊"
        positive += 1
    elif polarity < 0:
        sentiment = "Negative 😞"
        negative += 1
    else:
        sentiment = "Neutral 😐"
        neutral += 1
    print(f"\nReview {i+1}: {review[:45]}...")
    print(f"Score: {polarity:.2f} | {sentiment}")

print("\n" + "=" * 60)
print(f"SUMMARY: {positive} Positive | {negative} Negative | {neutral} Neutral")
print("=" * 60)
```
