```python
import dash
from dash import dcc, html
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

print("Dash ready! ✅")
```

```python
app = dash.Dash(__name__)

# Data for charts
projects = ['House Price', 'Customer Churn', 'Titanic', 'Neural Network', 'Sentiment', 'Recommender']
accuracy = [100, 78, 81, 58, 90, 85]

sentiment_data = pd.DataFrame({
    'Sentiment': ['Positive', 'Negative', 'Neutral'],
    'Count': [5, 3, 0]
})

titanic_data = pd.DataFrame({
    'Category': ['Female', 'Male', '1st Class', '2nd Class', '3rd Class'],
    'Survival Rate': [74, 19, 63, 47, 24]
})

app.layout = html.Div([
    html.H1("Kenny AI/ML Portfolio Dashboard",
            style={'textAlign': 'center', 'color': '#2c3e50', 'fontFamily': 'Arial'}),

    html.H3("Model Performance Overview",
            style={'textAlign': 'center', 'color': '#7f8c8d'}),

    dcc.Graph(
        figure=px.bar(
            x=projects, y=accuracy,
            labels={'x': 'Project', 'y': 'Accuracy (%)'},
            title='Model Accuracy Across Projects',
            color=accuracy,
            color_continuous_scale='Blues'
        )
    ),

    html.Div([
        html.Div([
            dcc.Graph(
                figure=px.pie(
                    sentiment_data,
                    values='Count',
                    names='Sentiment',
                    title='Sentiment Analysis Results',
                    color_discrete_map={
                        'Positive': 'green',
                        'Negative': 'red',
                        'Neutral': 'gray'
                    }
                )
            )
        ], style={'width': '50%', 'display': 'inline-block'}),

        html.Div([
            dcc.Graph(
                figure=px.bar(
                    titanic_data,
                    x='Category',
                    y='Survival Rate',
                    title='Titanic Survival Rates (%)',
                    color='Survival Rate',
                    color_continuous_scale='Reds'
                )
            )
        ], style={'width': '50%', 'display': 'inline-block'})
    ]),

    html.Footer("Built by Kenny | AI/ML Portfolio 2026",
                style={'textAlign': 'center', 'color': '#95a5a6', 'padding': '20px'})
])

print("Dashboard created! ✅")
```

```python
if __name__ == '__main__':
    app.run(debug=True, port=8050)
```
