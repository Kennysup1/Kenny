```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

print("Libraries ready! ✅")
```

```python
# Users and their movie ratings (1-5 stars)
ratings_data = {
    'User':  ['Alice', 'Alice', 'Alice', 'Alice',
              'Bob',   'Bob',   'Bob',   'Bob',
              'Carol', 'Carol', 'Carol', 'Carol',
              'David', 'David', 'David', 'David',
              'Eve',   'Eve',   'Eve',   'Eve'],
    'Movie': ['Inception', 'Titanic', 'Avengers', 'Interstellar',
              'Inception', 'Titanic', 'The Matrix', 'Interstellar',
              'Titanic',   'Avengers','The Matrix', 'Interstellar',
              'Inception', 'Avengers','The Matrix', 'Interstellar',
              'Titanic',   'Avengers','The Matrix', 'Inception'],
    'Rating':[5, 3, 4, 5,
              4, 2, 5, 4,
              5, 4, 3, 2,
              3, 5, 4, 5,
              4, 3, 5, 2]
}

df = pd.DataFrame(ratings_data)
movie_matrix = df.pivot_table(index='User', columns='Movie', values='Rating').fillna(0)
print(movie_matrix)
```

```python
similarity = cosine_similarity(movie_matrix)
similarity_df = pd.DataFrame(similarity, index=movie_matrix.index, columns=movie_matrix.index)
print("User similarity matrix:")
print(similarity_df.round(2))
```

```python
def recommend_movies(user, n=3):
    user_similarities = similarity_df[user].drop(user)
    most_similar = user_similarities.nlargest(2).index
    
    user_watched = set(df[df['User']==user]['Movie'])
    recommendations = {}
    
    for similar_user in most_similar:
        similar_movies = df[df['User']==similar_user]
        for _, row in similar_movies.iterrows():
            if row['Movie'] not in user_watched:
                if row['Movie'] not in recommendations:
                    recommendations[row['Movie']] = 0
                recommendations[row['Movie']] += row['Rating']
    
    recommended = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)[:n]
    return recommended

print("Movies recommended for Alice:")
for movie, score in recommend_movies('Alice'):
    print(f"  {movie} (score: {score})")

print("\nMovies recommended for Bob:")
for movie, score in recommend_movies('Bob'):
    print(f"  {movie} (score: {score})")
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(similarity_df, cmap='YlOrRd')
plt.colorbar(label='Similarity Score')
plt.xticks(range(len(similarity_df.columns)), similarity_df.columns, rotation=45)
plt.yticks(range(len(similarity_df.index)), similarity_df.index)
plt.title('User Similarity Matrix - Kenny Recommender')
plt.tight_layout()
plt.show()
```
