```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Alice', 'Bob', None, 'David', 'Eve', 'Frank'],
    'Age': [25, None, 30, 22, 35, None],
    'Salary': [50000, 60000, 55000, None, 70000, 65000],
    'City': ['New York', 'london', 'PARIS', 'New York', None, 'paris']
}
df = pd.DataFrame(data)
print(df)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df['Age'])
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print(df['Salary'])
```

```python
df['Name'] = df['Name'].fillna('Unknown')
df['City'] = df['City'].fillna('Unknown')
print(df)
```

```python
df['City'] = df['City'].str.title()
print(df)
```

```python
print("Missing values after cleaning:")
print(df.isnull().sum())
print("\nClean dataset:")
print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Salary'], color='steelblue')
plt.xlabel('Name')
plt.ylabel('Salary ($)')
plt.title('Salary by Person - Kenny Data Cleaning')
plt.grid(axis='y')
plt.show()
```
